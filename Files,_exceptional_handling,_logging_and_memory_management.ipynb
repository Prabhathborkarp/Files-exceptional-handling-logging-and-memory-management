####Q9. What is the difference between multithreading and multiprocessing?

####Answer:
In Python, both multithreading and multiprocessing are techniques used to perform multiple tasks simultaneously, but they differ in how they use system resources and execute tasks.

**Multithreading**:

Multithreading allows multiple threads to run within the same process.

- Threads share the same memory space.
- Used for lightweight and faster tasks.
- Best suited for I/O-bound operations such as file handling, web requests, or database access.


        Example:
        import threading

        def display():
            print("Thread running")

        t1 = threading.Thread(target=display)
        t1.start()

Advantages
- Faster communication between threads
- Less memory usage
- Efficient for I/O operations

Disadvantages
- Threads share memory, which may cause data conflicts
- In Python, true parallel execution is limited by the Global Interpreter Lock (GIL)

**Multiprocessing**

Multiprocessing uses multiple processes, where each process has its own memory space.

- Processes run independently.
- Best suited for CPU-bound tasks such as calculations and data processing.
- Can achieve true parallel execution.

        Example:
        from multiprocessing import Process

        def display():
            print("Process running")

        p1 = Process(target=display)
        p1.start()

Advantages
- True parallelism
- Better performance for heavy computations
- Safer because processes do not share memory directly

Disadvantages
- Higher memory usage
- Process creation is slower than threads

Multithreading is suitable for tasks involving waiting operations like file or network access, while multiprocessing is better for computationally intensive tasks requiring true parallel execution.

####Q12. What are the basic steps involved in exception handling in Python?

####Answer:

Exception handling in Python is used to detect and manage runtime errors so that the program can continue executing smoothly without crashing.

Basic Steps Involved in Exception Handling

1. Identify Risky Code Using try

    The code that may generate an exception is placed inside the try block.

2. Handle Exceptions Using except

    If an error occurs in the try block, control moves to the appropriate except block.

3. Execute Code if No Error Occurs Using else

    The else block runs only if no exception occurs.

4. Perform Cleanup Using finally

    The finally block executes whether an exception occurs or not. It is commonly used for cleanup operations such as closing files or releasing resources.


        Example:
        try:
            num = int(input("Enter a number: "))
            result = 10 / num

        except ZeroDivisionError:
            print("Cannot divide by zero")

        except ValueError:
            print("Invalid input")

        else:
            print("Result:", result)

        finally:
            print("Program ended")

Flow of Exception Handling
1. Code inside try block executes
2. If an exception occurs, matching except block executes
3. If no exception occurs, else block executes
4. finally block executes in all cases

The basic steps in exception handling involve placing risky code in a try block, handling errors with except, optionally using else for successful execution, and using finally for cleanup tasks.

####Q15. How does Python's garbage collection system work?

####Answer:

Python’s garbage collection system in Python automatically manages memory by removing unused objects from memory. This helps prevent memory leaks and improves program efficiency.

How Garbage Collection Works in Python

Python mainly uses two mechanisms:

1. Reference Counting
2. Garbage Collector for Cyclic References


**1. Reference Counting:**

Every object in Python has a reference count that tracks how many variables refer to it.

- When a reference is created, the count increases.
- When a reference is removed, the count decreases.
- If the reference count becomes zero, the object is automatically deleted.

        Example:

        a = [1, 2, 3]
        b = a

Here:

- The list object is referenced by both a and b.
- Reference count becomes 2.

        del a
        #Now the reference count decreases to 1.


        del b
        #The reference count becomes 0, so Python frees the memory.

**2. Cyclic Garbage Collection:**

Reference counting alone cannot handle circular references.

Example of Circular Reference

    list1 = []
    list2 = []

    list1.append(list2)
    list2.append(list1)

Here:

- list1 references list2
- list2 references list1

Even if both variables are deleted, the objects still reference each other, so their reference counts do not become zero.

Python’s garbage collector detects such cyclic references and removes them automatically.

**Garbage Collection Module:**

Python provides the gc module to control garbage collection.

    import gc
    gc.collect()

This manually triggers garbage collection.

Advantages of Garbage Collection
- Automatically frees unused memory
- Prevents memory leaks
- Reduces manual memory management
- Improves program stability

Python’s garbage collection system works mainly through reference counting and cyclic garbage collection, automatically reclaiming unused memory and helping programs run efficiently and reliably.

os.fork()

os.fork() creates a new child process by duplicating the current process.

Working
- After fork(), two processes run:
    - Parent process
    - Child process
- The child process gets a copy of the parent process memory.

        Example:

        import os

        pid = os.fork()

        if pid == 0:
            print("Child Process")
        else:
            print("Parent Process")

multiprocessing Module

The multiprocessing module provides a higher-level and portable way to create processes.

Example:

    from multiprocessing import Process

    def display():
        print("Child Process")

    p = Process(target=display)
    p.start()
    p.join()


Features
- Works on Windows, Linux, and macOS
- Easier to use and manage
- Provides process synchronization and communication tools
- Safer and more flexible


os.fork() is a low-level Unix-specific method for creating processes, while the multiprocessing module provides a simpler, safer, and cross-platform approach for process management in Python.

####Q23. What are the challenges associated with memory management in Python?

####Answer:

Memory management in Python is automatic, but it still has several challenges that can affect program performance and efficiency.

Challenges Associated with Memory Management in Python
1. High Memory Consumption

    Python objects generally consume more memory compared to lower-level languages because of additional metadata and dynamic typing.

2. Garbage Collection Overhead

    Python’s garbage collector continuously checks for unused objects, which may slightly reduce performance in large applications.

3. Circular References

    Objects referencing each other may not be immediately freed by reference counting alone.

Example

    a = []
    b = []

    a.append(b)
    b.append(a)

    #This creates a circular reference that requires garbage collection to remove.

4. Memory Leaks

    Although Python has automatic memory management, memory leaks can still occur due to:

    - Unreleased resources
    - Global variables
    - Improper object references
5. Fragmentation

    Frequent allocation and deallocation of memory may lead to memory fragmentation, reducing efficient memory usage.

6. Difficulty in Tracking Memory Usage

    In large applications, identifying which objects consume excessive memory can be challenging.

7. Performance Impact

    Automatic memory management may slow down performance for memory-intensive or real-time applications.

8. Delayed Resource Release

    Objects may not be destroyed immediately after becoming unused because garbage collection timing is not always predictable.

Example of Manual Garbage Collection

    import gc

    gc.collect()

    #This manually triggers garbage collection.

Ways to Reduce Memory Management Issues
- Use efficient data structures
- Remove unnecessary references
- Close files and connections properly
- Avoid creating excessive temporary objects
- Use generators for large datasets

Although Python provides automatic memory management through reference counting and garbage collection, challenges such as high memory usage, circular references, memory leaks, and performance overhead can still occur and must be managed carefully.

This situation can be handled using exception handling with try and except blocks.

```python
# Step 1: Use try-except block to handle file not found error
try:
    # Step 2: Try to open a file that doesn't exist
    with open("missing.txt", "r") as file:

        # Step 3: Read and print each line if file exists
        for line in file:
            print(line.strip())

# Step 4: Catch the FileNotFoundError if file doesn't exist
except FileNotFoundError:
    print("Error: File not found. Please check the filename.")

# Step 5: Finally block — executes regardless of error or not
finally:
    print("Program executed successfully.")
```

####Q8. Write a program to handle a file opening error using exception handling.

####Answer:

```python
import logging
import os

# Step 1: Create a logger with a file handler
logger = logging.getLogger("file_logger")
logger.setLevel(logging.ERROR)
handler = logging.FileHandler("/content/file_error.log")
handler.setFormatter(logging.Formatter("%(asctime)s - %(levelname)s - %(message)s"))
logger.addHandler(handler)

# Step 2: Define a function to open a file
def open_file(filename):
    try:
        # Step 3: Try to open the file
        with open(filename, "r") as file:
            print(f"--- {filename} Contents ---")
            for line in file:
                print(line.strip())

    # Step 4: Catch file not found error
    except FileNotFoundError:
        print(f"Error: '{filename}' not found.")
        logger.error(f"FileNotFoundError: '{filename}' not found.")

    # Step 5: Catch permission error
    except PermissionError:
        print(f"Error: No permission to open '{filename}'.")
        logger.error(f"PermissionError: No permission to open '{filename}'.")

    # Step 6: Catch any other unexpected error
    except Exception as e:
        print(f"Unexpected error: {e}")
        logger.error(f"Unexpected error: {e}")

    # Step 7: Finally block — executes regardless of error or not
    finally:
        print("File operation completed.\n")

# Step 8: Test with valid and invalid files
open_file("missing.txt")            # → file not found
open_file("another_missing.txt")    # → file not found

# Step 9: Create a valid file and test
with open("/content/valid.txt", "w") as file:
    file.write("Hello, my name is Prabhath.\n")
    file.write("I am learning Python.")

open_file("/content/valid.txt")     # → valid file

# Step 10: Read and print the log file to verify
print("--- file_error.log Contents ---")
with open("/content/file_error.log", "r") as file:
    for line in file:
        print(line.strip())
```

####Q12. Write a program that demonstrates using multiple except blocks to handle different types of exceptions.

####Answer:

```python
# Step 1: Define a function to demonstrate multiple exceptions
def perform_operations(a, b, data, index):

    # Step 2: Handle ZeroDivisionError
    try:
        result = a / b
        print(f"Division:        {a} / {b} = {result}")
    except ZeroDivisionError:
        print(f"Error: Cannot divide {a} by zero.")
    finally:
        print("Division operation completed.\n")

    # Step 3: Handle KeyError
    try:
        grade = data[index]
        print(f"Grade of {index}: {grade}")
    except KeyError:
        print(f"Error: '{index}' not found in dictionary.")
    finally:
        print("Dictionary operation completed.\n")

    # Step 4: Handle IndexError
    try:
        students = ["Prabhath", "Raj", "Priya"]
        print(f"Student:         {students[index]}")
    except IndexError:
        print(f"Error: Index '{index}' out of range.")
    except TypeError:
        print(f"Error: Index must be an integer, not '{type(index).__name__}'.")
    finally:
        print("List operation completed.\n")

    # Step 5: Handle ValueError
    try:
        number = int(index)
        print(f"Converted:       '{index}' to integer {number}")
    except ValueError:
        print(f"Error: Cannot convert '{index}' to integer.")
    finally:
        print("Conversion operation completed.\n")

# Step 6: Test with valid inputs
print("========== Valid Inputs ==========")
perform_operations(10, 2, {"Prabhath": 90, "Raj": 85}, "Prabhath")

# Step 7: Test with invalid inputs
print("========== Invalid Inputs ==========")
perform_operations(10, 0, {"Prabhath": 90, "Raj": 85}, "Arjun")
```

####Q14. Write a program that uses the logging module to log both informational and error messages.

####Answer:

```python
import logging

# Step 1: Create a logger
logger = logging.getLogger("app_logger")
logger.setLevel(logging.DEBUG)              # capture all levels

# Step 2: Create a file handler to write logs to a file
file_handler = logging.FileHandler("/content/app.log")
file_handler.setFormatter(logging.Formatter("%(asctime)s - %(levelname)s - %(message)s"))
logger.addHandler(file_handler)

# Step 3: Create a stream handler to print logs to console
console_handler = logging.StreamHandler()
console_handler.setFormatter(logging.Formatter("%(levelname)s - %(message)s"))
logger.addHandler(console_handler)

# Step 4: Define a function to perform student operations
def add_student(name, grade):
    try:
        # Step 5: Log informational message
        logger.info(f"Adding student: {name} with grade: {grade}")

        if not isinstance(grade, (int, float)):
            raise ValueError(f"Grade must be a number, not '{type(grade).__name__}'.")

        if grade < 0 or grade > 100:
            raise ValueError(f"Grade must be between 0 and 100, not {grade}.")

        # Step 6: Log success message
        logger.info(f"Student '{name}' added successfully with grade {grade}.")
        print(f"Student '{name}' added successfully.")

    # Step 7: Log error message
    except ValueError as e:
        logger.error(f"Failed to add student '{name}': {e}")
        print(f"Error: {e}")

    finally:
        print("Operation completed.\n")

# Step 8: Test with valid and invalid inputs
add_student("Prabhath", 90)         # → valid
add_student("Raj",      85)         # → valid
add_student("Priya",    105)        # → invalid grade — out of range
add_student("Arjun",    "A+")       # → invalid grade — wrong type
add_student("Sneha",    -10)        # → invalid grade — negative

# Step 9: Read and print the log file to verify
print("--- app.log Contents ---")
with open("/content/app.log", "r") as file:
    for line in file:
        print(line.strip())
```

```python
import os

# Step 1: Define a function to read and print file content
def read_file(filename):
    try:
        # Step 2: Check if file exists
        if not os.path.exists(filename):
            print(f"Error: '{filename}' does not exist.")
            return

        # Step 3: Check if file is empty
        if os.path.getsize(filename) == 0:
            print(f"Error: '{filename}' is empty.")
            return

        # Step 4: Open and print file content
        print(f"--- {filename} Contents ---")
        with open(filename, "r") as file:
            for line in file:
                print(line.strip())

    # Step 5: Catch any unexpected error
    except Exception as e:
        print(f"Unexpected error: {e}")

    finally:
        print("File operation completed.\n")

# Step 6: Create a file with content
with open("/content/students.txt", "w") as file:
    file.write("Prabhath\n")
    file.write("Raj\n")
    file.write("Priya\n")

# Step 7: Create an empty file
with open("/content/empty.txt", "w") as file:
    pass                                    # creates empty file

# Step 8: Test with valid, empty and missing files
read_file("/content/students.txt")          # → valid file
read_file("/content/empty.txt")             # → empty file
read_file("/content/missing.txt")           # → missing file
```
